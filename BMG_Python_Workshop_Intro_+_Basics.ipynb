**Debugging lists**

*Source: Think Python*

Careless use of lists can lead to long hours of debugging. Here are some pitfalls and how to avoid them:

1. Most list methods modify the argument and return `None`.

It is tempting to write list code like this:

```
t = t.sort()    # WRONG!
```

Because `sort` returns `None`, the next operation you perform with `t` is likely to fail.

2. Pick an idiom and stick with it.

Part of the problem with lists is that there are too many ways to do things.

For example, to add an element, you can use the `append` method or the + operator. Assuming that `t` is a list and `x` is a list element, these are correct:

```
t.append(x)
t = t + [x]
t += [a]
```

And these are wrong:

```
t. append([x])      # WRONG!
t = t.append(x)     # WRONG!
t + [x]             # WRONG!
t = t + x           # WRONG!
```

Try each of the example to make sure you understand what they do. Notice that only the last one causes a runtime error; the other three are legal, but they do the wrong thing.

3. Make copies to avoid aliasing

If you want to use a method like `sort` that modifies the argument, but you need to keep the original list as well, you can make a copy.

```
t = [3,1,2]
t2 = t[:]

t2.sort()
```

In this example, you couls also use the built-in function `sorted`, which returns a new, sorted list and leaves the original alone.

```
t2 = sorted(t)
```

**Conditional statement**

```
if condition:
  code to execute if condition is satisfied
else:
  code to exectute if condition is not statisfied
```

**Example (Conditional statement)**

````python
x = 3
y = 5

if x == y:
  print('x and y are equal')
else:
  print('x and y are not equal')
````

**Example (Nested conditionals)**

````python
x = 3
y = 5

if x == y:
  print('x and y are equal')
elif x < y:
  print('x is less than y')
elif x > y:
  print('x is greater than y')
````

````python
x = 3
y = 5

if x == y:
  print('x and y are equal')
else:
  if x < y:
    print('x is less than y')
  else:
    print('x is greater than y')
````

## Iteration: `while` loop

*   Enables you to repeat a operation while a certain condition is satisfied.

```
while condition:
  code to be executed while condition is satified
```

**Example**

````python
n = 10

while n > 0:
  print(n)
  n = n - 1
````

**Exercise**

1.  Set a variable called `sum` to 0.
2.  Select a random value between 1 and 6 (to simulate a throw of a die).
3.  Add the random value from step 2 to the `sum` variable
4.  Repeat steps 2 and 3 until the value of `sum` is greater than or equal to 50.



````python
sum = 0

while sum < 50:
  value = np.random.randint(1,7,1)
  sum = sum + value
  print(sum)
````
