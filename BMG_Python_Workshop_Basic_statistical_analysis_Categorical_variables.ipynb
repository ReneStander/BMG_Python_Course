# Contingency tables & Probability
## Example: Jerusalema

A researcher was interested in whether animals could be trained to line dance. She took 80 cats and tried to train them to line dance by giving them either food or love as a reward. At the end of the week, she counted how many cats could line dance and how many could not.

Consider the following 2-by-2 contingency table:

JerusalemaContingencyTable.png



Import the necessary packages.

```python
import numpy as np
from scipy.stats import chi2_contingency
```

Give the contingency table with **observed** counts.

```python
ct_cats = np.array([[28, 12], [16, 24]])
print(ct_cats)
```

One of the cats is randomly selected. Consider the following two events:

*A* : The cat could line dance.

*B* : The cat was given food as reward.

What is the probability that the cat could line dance?

```python
prob_A = (28 + 16) / 80
print(prob_A)
```

What is the probability that the cat could not line dance?

```python
prob_Acomp = 1 - prob_A
print(prob_Acomp)
```

What is the probability that the cat's reward was food?

```python
prob_B = (28 + 12) / 80
print(prob_B)
```

What is the probability that the cat could line dance and that
the cat's reward was food?

```python
prob_AandB = 28 / 80
print(prob_AandB)
```

What is the probability that the cat could line dance given that
the cat's reward was food?

```python
prob_AgivenB = 28 / 40
print(prob_AgivenB)
```

Does learning to line dance depend on the type of reward?

```python
print(prob_A)
print(prob_AgivenB)
prob_A == prob_AgivenB
```

Therefore *A* and *B* are not independent events.

In effect, learning to line dance depends on the type of award.

# Testing for independence
## Example: Jerusalema

Use Pearson's chi-square test to test whether the type of reward
and learning to line dance are independent.

$H_{0}:$ Type of reward and learning to line dance are **independent**.

$H_{A}:$ Type of reward and learning to line dance are **dependent**.

Compare the observed counts with the expected counts.

Observed counts: $O_{i,j}$

Expected counts: $E_{i,j}=\frac{i^{th}\text{ row total }\times \text{ }j^{th}\text{ column
total}}{n}$

Test statistic: $\chi ^{2}=\sum_{i}\sum_{j}\frac{\left( O_{i,j}-E_{i,j}\right) ^{2}}{E_{i,j}}
$


```python
chi2_contingency(ct_cats)
```

Since *p*-value < 0.05, $H_{0}$ is rejected at a 5% significance level. Type of reward and learning to line dance are dependent.
