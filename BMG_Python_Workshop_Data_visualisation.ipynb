## Example: South African birds

Consider the following variables for 108 animals:

*   Card number
*   Type of animal
*   Habitat
*   Species
*   Age in years
*   Weight in kilograms
*   Size in centimeters
*   Speed in kilometer per hour
*   Vulnerability status scaled from 1 (highest risk) to 4 (lowest risk)

The focus will be on the 36 birds in the data set.

Import the necessary packages.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

Import the data set.

```python
animals = pd.read_csv("SuperAnimals.csv")
animals
```

Subset the data to only consider the birds.

```python
birds = animals.loc[animals["Species"] == "Bird", :]
birds
```

Draw a barplot showing the number of birds in the different habitats.

```python
birds_habitat = birds["Habitat"].value_counts()
sns.barplot(x = birds_habitat.index, y = birds_habitat.values)
```

Draw a barplot showing the vulnerability status of the birds.

```python
birds_vulnerability = birds["Vulnerability"].value_counts()
sns.barplot(x = birds_vulnerability.index, y = birds_vulnerability.values)
```

Draw a boxplot for the age in years of the birds.

```python
box_Age = sns.boxplot(data = birds,
                      x = "Age",
                      color = "lightgreen")
plt.show()
```

Draw a boxplot for the size in cm of the birds.

```python
box_Size = sns.boxplot(data = birds,
                       x = "Size",
                       color = "lightblue")
plt.show()
```

Draw a boxplot for the speed of the birds in km per hour.

```python
box_Speed = sns.boxplot(data = birds,
                        x = "Speed",
                        color = "orange")
plt.show()
```

Draw a boxplot for the weight in kg of the birds.

```python
box_Weight = sns.boxplot(data = birds,
                         x = "Weight",
                         color = "pink")
plt.show()
```

Apply log transformations to the size and to the weight of the birds

```python
birds["logSize"] = np.log(birds["Size"])
birds["logWeight"] = np.log(birds["Weight"])
```

Draw a boxplot for the log-transformed size of the birds.

```python
box_logSize = sns.boxplot(data = birds,
                          x = "logSize",
                          color = "lightblue")
plt.show()
```

Draw a boxplot for the log-transformed weight of the birds.

```python
box_logWeight = sns.boxplot(data = birds,
                            x = "logWeight",
                            color = "pink")
plt.show()
```

Draw a histogram for the size in cm of the birds.

```python
finish = sns.histplot(data = birds,
                      x = "Size",
                      color = "lightblue",
                      edgecolor = "blue",
                      bins = 8)
finish.set(xlabel = "Size in cm",
           ylabel = "Frequency",
           title = "Histogram of the size in cm of the birds")
plt.show()
```

Draw a histogram for the weight in kg of the birds.

```python
finish = sns.histplot(data = birds,
                      x = "Weight",
                      color = "pink",
                      edgecolor = "purple",
                      bins = 8)
finish.set(xlabel = "Weight in kg",
           ylabel = "Frequency",
           title = "Histogram of the weight in kg of the birds")
plt.show()
```

Draw a histogram for the log-transformed size of the birds.

```python
finish = sns.histplot(data = birds,
                      x = "logSize",
                      color = "lightblue",
                      edgecolor = "blue",
                      bins = 8)
finish.set(xlabel = "log(Size) where size is in cm",
           ylabel = "Frequency",
           title = "Histogram of the log-transformed size of the birds")
plt.show()
```

Draw a histogram for the log-transformed weight of the birds.

```python
finish = sns.histplot(data = birds,
                      x = "logWeight",
                      color = "pink",
                      edgecolor = "purple",
                      bins = 8)
finish.set(xlabel = "log(Weight) where weight is in kg",
           ylabel = "Frequency",
           title = "Histogram of the log-transformed weight of the birds")
plt.show()
```

Draw a scatterplot of the weight of the birds against the size of the birds

```python
scatt = sns.scatterplot(data = birds,
                        x = "Size",
                        y = "Weight",
                        color = "red")
scatt.set(xlabel = "Size in cm",
          ylabel = "Weight in kg")
plt.show()
```

Draw a scatterplot of the log-transformed weight of the birds against the log-transformed size of the birds.

```python
scatt = sns.scatterplot(data = birds,
                        x = "logSize",
                        y = "logWeight",
                        color = "red")
scatt.set(xlabel = "log(Size) where size is in cm",
          ylabel = "log(Weight) where weight is in kg")
plt.show()
```
